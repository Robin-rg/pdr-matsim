```python
import xml.etree.ElementTree as ET

# Chargement du fichier XML
tree = ET.parse('E:/Code/MATSim/pdr-matsim/scenarios/douai/douai_population.xml')
root = tree.getroot()

# Parcourir toutes les personnes dans le fichier
for person in root.findall(".//person"):
    person_id = person.get("id")  # Obtenir l'ID de la personne

    # Trouver l'attribut 'vehicles' pour cette personne
    for attribute in person.findall(".//attributes/attribute[@name='vehicles']"):
        # Récupérer la valeur actuelle de l'attribut 'vehicles'
        vehicles_value = attribute.text.strip()

        # Vérifier si la personne possède un véhicule "car"
        if '"car"' in vehicles_value:
            # Ajouter le véhicule
            drs_driver = f'"drsDriver":"{person_id}:car"' # !!! à tester

            # Si "drsDriver" n'est pas déjà présent, l'ajouter
            if '"drsDriver"' not in vehicles_value:
                vehicles_value = vehicles_value.rstrip('}') + f', {drs_driver} }}'

                # Mettre à jour l'attribut 'vehicles' avec la nouvelle valeur
                attribute.text = vehicles_value

# Sauvegarder le fichier modifié
tree.write('population_modified.xml', encoding='utf-8', xml_declaration=True)

print("Le véhicule 'drsDriver' a été ajouté pour les personnes possédant un véhicule 'car'.")
print("\nATTENTION LE SCRIPT RETIRE LA LIGNE DOCTYPE population !!! \nil faut la remettre")
```
